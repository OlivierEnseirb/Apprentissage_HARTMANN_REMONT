Une telle mthode n'est pas utilisable en l'tat pour l'apprentissage car comme son nom l'indique, elle joue avec l'alatoire, et en plus elle n'a pas de retour sensorimoteur pour connatre la position obtenue. En effet,  chaque nouvelle configuration moteur une nouvelle position est obtenue et peu ne rien  voir avec la prcdente. Le robot ne peut pas faire le lien entre telle modification de "position articulaire d'un coude" va avoir tel "impact". Et cela est d'autant plus vrai que la dimention du bras augmente : les modles gomtriques direct et indirect en deviennent plus compliqus. Il serait certainement possible de les construire avec une telle mthode, mais cela serait beaucoup trop long et fastidueux.
