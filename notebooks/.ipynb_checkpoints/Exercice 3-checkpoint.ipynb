Etude du modèle random goal babbling nearest_neighbor:

```python
from explauto.experiment import Experiment, make_settings
from explauto.environment import environments


s = make_settings(environment='simple_arm',
                  babbling_mode='goal', 
                  interest_model='random',
                  sensorimotor_model='nearest_neighbor',
                  environment_config='default')

goal_xp = Experiment.from_settings(s)

goal_xp.evaluate_at([1, 10, 20, 30, 100, 200, 300, 400], s.default_testcases)
goal_xp.run()
```

```python
%pylab inline
ax_goal = axes()
data = goal_xp.log.scatter_plot(ax_goal, (('sensori', [0, 1]), ), color='green')
legend(('goal',))
```

```python
ax = axes()
goal_xp.log.plot_learning_curve(ax)

legend(('goal',))
```

La methode nearest_neighbor atteint les mêmes performances que la même méthode en motor babbling en aussi peu de temps et tend même plus rapidement vers 0.

Etude du modèle random goal babbling WNN:

```python
from explauto.experiment import Experiment, make_settings
from explauto.environment import environments


s = make_settings(environment='simple_arm',
                  babbling_mode='goal', 
                  interest_model='random',
                  sensorimotor_model='WNN',
                  environment_config='default')

goal_xp = Experiment.from_settings(s)

goal_xp.evaluate_at([1, 10, 20, 30, 100, 200, 300, 400], s.default_testcases)
goal_xp.run()
```

```python
%pylab inline
ax_goal = axes()
data = goal_xp.log.scatter_plot(ax_goal, (('sensori', [0, 1]), ), color='green')
legend(('goal',))
```

```python
ax = axes()
goal_xp.log.plot_learning_curve(ax)

legend(('goal',))
```

Encore une fois, l'erreur tombe très rapidement de très haut pour atteindre des valeurs raisonnables mais moins vite que la précédente, et moins lentement que la première. Cependant, celle ci stagne ensuite à 0.1m d'erreur.

Ces trois méthodes donnent des résultats significatifs très rapidement de façon exponentielle, et donc les erreurs convergent vers 0, excepté pour la dernière méthode qui la voit stagner à une valeur assez importante par rapport aux capacités des autres. Par ailleurs, la méthode LWLR-CMAES semble plus efficace que la nearest_neighbor.
