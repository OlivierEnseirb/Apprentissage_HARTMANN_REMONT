Interprétation:
Pour chacune de ces configurations on atteint loin de là l'ensemble de l'espace de travail du robot mais les points atteints représentent plutôt un arc de cercle. De plus, la dimentionnalité du bras n'y change rien, les résultats sont les "mêmes".
Une telle méthode n'est pas utilisable en l'état pour l'apprentissage car comme son nom l'indique, elle joue avec l'aléatoire, et en plus elle n'a pas de retour sensorimoteur pour connaître la position obtenue. En effet, à chaque nouvelle configuration moteur une nouvelle position est obtenue et peu ne rien à voir avec la précédente. Le robot ne peut pas faire le lien entre telle modification de "position articulaire d'un coude" va avoir tel "impact". Et cela est d'autant plus vrai que la dimention du bras augmente : les modèles géométriques direct et indirect en deviennent plus compliqués. Il serait certainement possible de les construire avec une telle méthode, mais cela serait beaucoup trop long et fastidueux.
