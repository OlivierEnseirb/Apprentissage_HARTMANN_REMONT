Etude du mod√®le random goal babbling nearest_neighbor:

```python
from explauto.experiment import Experiment, make_settings
from explauto.environment import environments


s = make_settings(environment='simple_arm',
                  babbling_mode='goal', 
                  interest_model='random',
                  sensorimotor_model='nearest_neighbor',
                  environment_config='default')

goal_xp = Experiment.from_settings(s)

goal_xp.evaluate_at([1, 10, 20, 30, 100, 200, 300, 400], s.default_testcases)
goal_xp.run()
```

```python
%pylab inline
ax_goal = axes()
data = goal_xp.log.scatter_plot(ax_goal, (('sensori', [0, 1]), ), color='green')
legend(('goal',))
```

```python
ax = axes()
goal_xp.log.plot_learning_curve(ax)

legend(('goal',))
```

La methode nearest_neighbor atteint les m√™mes performances que la m√™me m√©thode en motor babbling en aussi peu de temps et tend m√™me plus rapidement vers 0.
